### After training the model for 30 epochs for each of the p number of ranks in problem one,  let’s  compare  both  our  models’  accuracy  and  their  total  runtime  for  each  run. Create two plots:
#### a.  A line plot showing the epoch accuracy for each of the runs
#### b.  A second line plot that shows the total runtime by the number of processes used

We will use matlotlib and pandas to create these plots:

```python
import matplotlib.pyplot as plt
import pandas as pd
```

Then, we import all the results using pandas:

```python
metrics_p1 = pd.read_csv("output/metrics_p1.csv")
metrics_p2 = pd.read_csv("output/metrics_p2.csv")
metrics_p4 = pd.read_csv("output/metrics_p4.csv")
metrics_p8 = pd.read_csv("output/metrics_p8.csv")
```

The following figure is to reproduce results from the lecture showing loss:

```python
plt.figure()
plt.plot(metrics_p1.elapsed_time, metrics_p1.epoch_losses,label="P=1",marker="o")
plt.plot(metrics_p2.elapsed_time, metrics_p2.epoch_losses,label="P=2",marker="o")
plt.plot(metrics_p4.elapsed_time, metrics_p4.epoch_losses,label="P=4",marker="o")
plt.plot(metrics_p8.elapsed_time, metrics_p8.epoch_losses,label="P=8",marker="o")
plt.legend()
plt.title("MNIST Loss")
plt.xlabel("Elapsed Time")
plt.ylabel("Loss")
plt.show()
```

The following figure shows the accuracy over time (addresses Problem 2.a):

```python
plt.figure()
plt.plot(metrics_p1.elapsed_time, metrics_p1.epoch_accuracies,label="P=1",marker="o")
plt.plot(metrics_p2.elapsed_time, metrics_p2.epoch_accuracies,label="P=2",marker="o")
plt.plot(metrics_p4.elapsed_time, metrics_p4.epoch_accuracies,label="P=4",marker="o")
plt.plot(metrics_p8.elapsed_time, metrics_p8.epoch_accuracies,label="P=8",marker="o")
plt.legend()
plt.title("MNIST Accuracy")
plt.xlabel("Elapsed Time")
plt.ylabel("Accuracy (%)")
plt.show()
```

The following figure shows the total runtime by the number of processors used (addresses Problem 2.b):

```python
plt.figure()
#plt.plot([1,2,4,8],[metrics_p1.elapsed_time[29],metrics_p2.elapsed_time[29],metrics_p4.elapsed_time[29],metrics_p8.elapsed_time[29]],marker="o")
plt.plot(["1","2","4","8"],[metrics_p1.elapsed_time[29],metrics_p2.elapsed_time[29],metrics_p4.elapsed_time[29],metrics_p8.elapsed_time[29]],marker="o")
plt.xlabel("Number of Processes")
plt.ylabel("Runtime (s)")
plt.show()
```

### What is the effect of training time as we increase the number of processes available?
The results show that training for this data set benefits from using additional processes. While the highest accuracy was achieved by the single process case, it has the longest runtime. The gain in accuracy comes at an expense significant increase in runtime. Using 4 processes, we see comparable accuracy at approximately a third of the runtime.

### How does this line up with your expectations of the scalability laws that we discussed?
The results show what you would expect. The problem is not embarrasingly parallel, therefore you see less than linear speedup as you increase the number of processes. 

### Which  scalability  law  is  most  applicable  in  this  case?
As seen in class and in this data parallelism assignment, Gustaffson's Law is most applicable.

### How  is  our  model  accuracy affected by the increase in the number of processes running? 
Accuracy significantly decreases between 1 and 2 processes, but using 4 and 8 processes, we are able to achieve comparable accuracy as that shown in the serial case.
